```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('errors_by_path_table.csv')
print(df.head())
```

```python
error_counts = pd.DataFrame({
    'Error Type': ['404 Errors', '500 Errors'],
    'Count': [df['404_errors'].sum(), df['500_errors'].sum()]
})
```

```python
error_counts.set_index('Error Type').plot(kind='bar', legend=False)
plt.title('Total Number of 404 vs 500 Errors')
plt.xlabel('Error Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# 2. Identify the top five pages with 500 errors
top_500_errors = df.nlargest(5, '500_errors')[['path', '500_errors']]

# Plotting
top_500_errors.set_index('path').plot(kind='bar')
plt.title('Top Five Pages with 500 Errors')
plt.xlabel('Page')
plt.ylabel('Count of 500 Errors')
plt.xticks(rotation=45)
plt.show()
```
